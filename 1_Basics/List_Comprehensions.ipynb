```python
#Generating numbers from 0 to 100

numbers = [x for x in range(100)]

numbers
```

```python
numbers = []

for number in range(100):
  numbers.append(number)
```

```python
numbers
```

```python
numbers = [x*2 for x in range(100)]

numbers
```

```python
#We can convert x to a float

numbers = [float(x) for x in range(100)]

numbers
```

```python
value = [x for x in 'PYTHON']

value
```

```python
#We can also do this in form of a set

value = {x for x in 'PYTHON'}

value
```
