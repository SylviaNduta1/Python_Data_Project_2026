**Median Salary vs Data Science Jobs**

```python
!pip install datasets

# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Get the average salary by job title
job_salary = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
job_salary
```

```python
job_salary = df.groupby('job_title_short')['salary_year_avg'].mean().sort_values()
job_salary
```

```python
# Plot the chart
job_salary.plot(kind='barh')

# Customize the chart
plt.xlabel('Salary ($USD)')
plt.ylabel('')
plt.title('Median Salary by Job Title')

# Show the bar chart
plt.show()
```
